## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
#Ejemplo 90 grados
grados = float (input("Dame un valor en grados:"))
pi = 3.1416
radianes = grados *(pi/180)
round_radianes = round(radianes,2)
print("Este es el resultado, radianes = ",round_radianes, "para los", grados, "º introducidos")
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
#Ejemplo base=5 y altura = 5
base = int (input("Introduzca medida de la base en m"))
altura = int (input("Introduzca medida de la altura en m"))
area = base * altura
print("Con los datos insertados obtenemos un area de:",area,"m2")
```
