## PARTE I

En este ejercicio debes simular un chatbot al que le haremos pedidos de pizzas.

1. El chatbot tiene que saludar con un: "Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana ¿Cuántas pizzas  familiares desea?"
2. El ususario tiene que introducir un número de pizzas en una variable llamada 'pizz_familiar'
3. El chatbot debe responder: "Estupendo, ¿cuántas pizzas medianas desea?" y guardar el número de pizzas medianas en otra variable 'pizz_mediana'
4. Además debe responder del chatbot: "Estupendo, se están preparando 'pizz' pizzas familiares y 'pizz_mediana' pizzas medianas. Digame su dirección"
5. El ususario tiene que introducir una direccion en formato String en otra variable llamada 'direcc'
6. El programa debe calcular el monto total
7. Respuesta final del chatbot: "Le mandaremos las '(número total de pizzas)' pizzas a la dirección 'direcc'. Serán <precio total a pagar en euros>. Muchas gracias por su pedido."
         


```python
#He pedido 5 familiares y 6 medianas para la calle Manuel Garcia 7

pizz_familiar = input ("Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana ¿Cuántas pizzas  familiares desea?")
pizz_mediana = input ("Estupendo, ¿cuántas pizzas medianas desea?")
num_pizzas = int(pizz_familiar) + int(pizz_mediana)
direcc = input ("Estupendo, se están preparando pizzas familiares y pizzas medianas. Digame su dirección")
precio1=8.95 ; precio2=9.90
precio_total_a_pagar_en_euros = float(pizz_familiar)*precio1 + float(pizz_mediana)*precio2
print("Le mandaremos las",num_pizzas," pizzas a la dirección",direcc,". Serán",precio_total_a_pagar_en_euros,"€. Muchas gracias por su pedido.")
```

## Parte II

Crea un programa que realice las siguientes acciones:

1. Solicite al usuario que introduzca 3 números enteros separados por comas (por ejemplo: 1,2,3).
2. Convierta esa entrada en una lista de números enteros.
3. Utilizando funciones built-in, determine el número máximo, el número mínimo y la suma total de los números introducidos.
4. Determine la verdad o falsedad de la siguiente afirmación: "El número máximo menos el número mínimo es igual a la suma total dividida por 5".
5. Muestre los resultados utilizando print.

```python
#Número utilizados 3, 5, 17
usuario = input("introduzca 3 números enteros separados por comas") #1
lista = [int(num.strip()) for num in usuario.split(",")] #Se trata de un bucle for, para cada numero que entra eliminar las comas para que no las recoja la tabla
print("2. La lista es:",lista)                           #.strip es un metodo para eliminar huecos y .split para eliminar caracteres especificos. 

num_max = max(lista) ; num_min = min(lista) ; num_sum = sum(lista)
print("3. El número máximo:",num_max,", el número mínimo:",num_min," y la suma de la lista es:",num_sum)

if num_max - num_min == num_sum/5:
    print("4. El número máximo menos el número mínimo es igual a la suma total dividida por 5")
else:
    print("4. El número máximo menos el número mínimo NO es igual a la suma total dividida por 5")


#"El número máximo menos el número mínimo es igual a la suma total dividida por 5"
```
