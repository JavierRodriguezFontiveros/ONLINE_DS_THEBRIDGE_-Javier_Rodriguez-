## Ejercicio 10
Convierte el programa del ejercicio 2 en una función al que se le pase el valor a buscar como un argumento posicional

`````python
lista_palabras = ["un", "es", "binario", "es", "un", "otro", "binario", "un", "es"]

def repite(lista, repetida):
    indices = []  # Aquí se acumulan los indices
    for indice, elemento in enumerate(lista):  #enumerate() obtiene el indice y elemento
        if elemento == repetida:
            indices.append(indice) #Si el elemento se repite añade un indice a la lista y así hasta que no se repita más
    return indices

var_dup1 = repite(lista_palabras, "un") ; var_dup2 = repite(lista_palabras, "es") ; var_dup3 = repite(lista_palabras, "binario") #Llamamos alas funciones

print("Los índices de 'un':", var_dup1,", los índices de de'es:",var_dup2,"y los índices de 'binario':", var_dup3)

`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
dni = input("Introduce tu documento nacional de identificación con su letra al final sin guion: ") #47347494Q

def dni_valid(dni): #Función para comprobar que cumple el numero maximo y minimo de digitos + letra 
    palabra = "TRWAGMYFPDXBNJZSQVHLCKE"
    if len(dni) < 8 or len(dni) > 9:
        return False
    
    numero = dni[:-1]  #Diferencia las partes del dni introducido ya que se trata de un 'str'
    letra = dni[-1]  

    if not numero.isdigit(): #Comprueba si la parte numérica es valida
        return False
    
    numero = int(numero) 

    letra_correcta = palabra[numero % 23] 
    letra = letra.upper() #Daigual si se introduce la letra en minusculas o mayusculas
    return letra == letra_correcta 

if dni_valid(dni):
    print("El DNI es correcto: TRUE")
else:
    print("El DNI no es correcto: FALSE")


`````
