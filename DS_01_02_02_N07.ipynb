## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
#Tiendas de electrodomésticos
#Creamos la clase 'Tienda' y todos los metodos dentro de la clase

class Tienda:
    def __init__(self, nombre, direccion, n_empleados, ventas): #Atributos variables
        self.tipo = "Electrodomésticos" #Atributo fijo
        self.abierta = True #Atributo fijo
        self.nombre = nombre
        self.direccion = direccion
        self.n_empleados = n_empleados
        self.ventas = ventas

    #Creamos todos los metodos
    def calcular_ventas(self):
        return sum(self.ventas)
    
    def media_ventas_empleado(self):
        ventas_totales = self.calcular_ventas()
        if self.n_empleados > 0:
            return ventas_totales / self.n_empleados
        else:
            return 0
    
    def ubicacion(self):
        return print("Tienda:",self.nombre, "en la Dirección:",self.direccion)

    def ventas_ultimo_mes(self):
        if self.ventas:
            return self.ventas[-1]
        else:
            return 0

    def proyeccion_ventas(self, inversion):
        if inversion >= 1000:
            factor = 1.5
        else:
            factor = 1.2
        self.ventas = [venta * factor for venta in self.ventas]
        return self.ventas


#Creando 3 tiendas
tienda_JavierSA = Tienda("JavierSA", "Avenida Tomares", 600, [5250, 7500, 6400])
tienda_BlancaSA = Tienda("BlancaSA", "Calle Bormujos", 250, [3000, 5000, 4000])
tienda_AlfredoSA = Tienda("AlfredoSA", "Manzana Mairena", 120, [3500, 2600, 4200])

#Lista de tiendas
tiendas = [tienda_JavierSA, tienda_BlancaSA, tienda_AlfredoSA]

#Ventas del último mes en todas las tiendas
for tienda in tiendas: 
    print("Ventas en el último mes en", tienda.nombre , ":", tienda.ventas_ultimo_mes())

for tienda in tiendas:
    if "Avenida" in tienda.direccion:
        print("Tienda en la avenida:", tienda.nombre) #Se muestra solo la que contiene 'avenida'
        
#Probamos los métodos con la tienda de BlancaSA
print(tienda_BlancaSA.ubicacion())
print("Ventas totales últimos 3 meses:", tienda_BlancaSA.calcular_ventas())
print("Media de ventas por empleado:", tienda_BlancaSA.media_ventas_empleado())
print("Ventas del último mes:", tienda_BlancaSA.ventas_ultimo_mes())
print("Proyección de ventas con inversión de 1200:", tienda_BlancaSA.proyeccion_ventas(1200))

```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
#Programar a un perro
#Creamos la clase 'Perro' y todos los metodos dentro de la clase

class Perro:
    def __init__(self, raza, velocidad = 0, pelaje = "Marron", dueño = False): #Atributos variables
        self.patas = 4 #Atributo fijo
        self.orejas = 2 #Atributo fijo
        self.ojos = 2 #Atributo fijo
        self.velocidad = velocidad
        self.raza = raza
        self.pelaje = pelaje
        self.dueño = dueño
    
    #Cramos los métodos:
    def andar(self,aumento_de_velocidad):
        self.velocidad = self.velocidad + aumento_de_velocidad
        return self.velocidad
        
    def parar(self):
        self.velocidad = 0
        return self.velocidad

    def ladrar(self, entrada):
        return "GUAU!" + entrada
    
#Creamos al perrito Cheis
perro_cheis = Perro("Perro de aguas")

#Documentamos la clase de perro:
print("El perrito Cheis es un:", perro_cheis.raza , "y tiene el pelaje:", perro_cheis.pelaje)
print("Tiene:", perro_cheis.patas,"patas,", perro_cheis.orejas ,"orejas y", perro_cheis.ojos ,"ojos.")
print("Su velocidad actual es:", perro_cheis.velocidad)

#Probamos los métodos definidos:
print("El perrito Cheis se pone a andar:", perro_cheis.andar(10), ", a la velocidad de:",perro_cheis.velocidad)
print("El perrito Cheis se ha parado:", perro_cheis.parar()) 
print("El perrito Cheis está ladrando:", perro_cheis.ladrar("GUAAAAAAAU!")) 

```
